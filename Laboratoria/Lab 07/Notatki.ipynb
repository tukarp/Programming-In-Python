## Fraktale

### Fraktale 1

```python
# Funkcja generująca fraktale
def julia(z, c):
    for i in range(100):
        z = z * z + c
        if abs(z) > 2:
            return False
    return True

# Zaalokowanie zmiennych
c = -0.1 + 0.65j    # Wzór na c
d = 300             # Ilość punktów

# Punkty x i y
x = []
y = []

# Generacja fraktali
for i in range(d):
    for j in range(d):
        # Obliczanie punktów
        xi = -1.5 + ((3 / d) * i)
        yi = -1.5 + ((3 / d) * j)
        # Sprawdzenie czy punkt powinien zostać dodany
        if julia(xi + yi * 1j, c):
            # Dodanie punktów do list
            x.append(xi)
            y.append(yi)

# Rysowanie wykresu fraktali
plt.figure(figsize=(10, 10))
plt.scatter(x, y, c=y, s=1, cmap="jet")
plt.show()
```

### Fraktale 2

```python
# Funkcja generująca fraktale
def julia(z, c):
    for i in range(500):
        z = z * z + c
        if abs(z) > 2:
            return i
    return 100

# Zaalokowanie zmiennych
c = -0.4 + 0.65j    # Wzór na c
d = 300             # Ilość punktów
color = []          # Kolor

# Punkty x i y
x = []
y = []

# Generacja fraktali
for i in range(d):
    for j in range(d):
        # Obliczanie punktów
        xi = -1.5 + ((3 / d) * i)
        yi = -1.5 + ((3 / d) * j)
        # Obliczanie koloru
        color.append(julia(xi + yi * 1j, c))
        # Dodanie punktów do list
        x.append(xi)
        y.append(yi)

# Rysowanie wykresu fraktali
plt.figure(figsize=(10, 10))
plt.scatter(x, y, c=color, s=1, cmap="turbo")
plt.show()
```

### Fraktale 3

```python
# Funkcja generująca fraktale
def julia(z, c):
    for i in range(150):
        z = z * z + c
        if abs(z) > 2:
            return i
    return 100

# Zaalokowanie zmiennych
c = -0.8 + 0.156 * 1j        # Ilość punktów

# Generacja przestrzeni punktów x i y
X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 700), np.linspace(-1.5, 1.5, 700))

# Generacja siatki
Z = X + Y * 1j

# Generacja koloru
color = np.frompyfunc(julia, 2, 1)(Z, c).astype(np.float64)  # Kolor

# Rysowanie wykresu fraktali
plt.figure(figsize=(10, 10))
plt.contourf(X, Y, color, 200, levels = np.linspace(0, 100, 101), cmap="jet")
plt.show()
```

### Fraktale 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Funkcja generująca fraktale
def julia(z, c):
    for i in range(200):
        z = z * z + c
        if abs(z) > 2:
            return i
    return 200

# Generacja przestrzeni punktów x i y
X, Y = np.meshgrid(np.linspace(-2.5, 1.5, 500), np.linspace(-2.5, 1.5, 500))

# Generacja siatki
Z = X + Y * 1j
c = X + Y * 1j

# Generacja koloru
color = np.frompyfunc(julia, 2, 1)(Z, c).astype(np.float64)  # Kolor

# Rysowanie wykresu fraktali
plt.figure(figsize=(10, 10))
plt.contourf(X, Y, color, 200, levels = np.linspace(0, 100, 101), cmap="jet")
plt.show()
```
