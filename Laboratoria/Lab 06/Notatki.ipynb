### Porównanie czasów operacji w języku Python i bibliotece NumPy

```python
# Zaalokowanie zmiennych
xin = range(10**6)
xout = []

# Czas 1
t = time()
for i in xin:
    xout.append(i**2)
print(f"Czas 1: {time() - t}")

# Czas 2
t = time()
xout = [i**2 for i in xin]
print(f"Czas 2: {time() - t}")

# Czas 3
t = time()
xout = list(map(lambda i : i**2, xin))
print(f"Czas 3: {time() - t}")

# Czas 4
t = time()
xout = list(map((2).__rpow__, xin))
print(f"Czas 4: {time() - t}")

# Czas 4
t = time()
xin = np.arange(10**6)
xout = xin**2
print(f"Czas 5: {time() - t}")
```

### Wykresy

```python
print("Wykres 1:")
x = [i for i in range(0, 11)]
y = [i**2 for i in x]
plt.plot(x, y, marker="o", markersize=5, linewidth=2)
```

```python
print("Wykres 2:")
pi = np.pi
x = [-pi + 2 * pi / 100 * i for i in range(101)]
y = [math.cos(i) for i in x]
plt.plot(x, y)
```

```python
print("Wykres 3:")
x = np.linspace(-pi, pi, 100)
plt.plot(x, np.exp(x), label="$e^x$")
plt.legend()
plt.show()
```

```python
print("Wykres 4:")
x = np.linspace(0, 1.0, 101)
for i in np.linspace(0.25, 2, 8):
    plt.plot(x, x**i, color=(i /2, 0, 1 - i / 2))
```

```python
print("Wykres Zadania:")
array = np.array([
    [[0.85, 0.04, 0], [-0.04, 0.85, 1.6]],
    [[-0.15, 0.28, 0], [0.26, 0.24, 0.44]],
    [[0.2, -0.26, 0], [0.23, 0.22, 1.6]],
    [[0, 0, 0], [0, 0.16, 0]]
])

x = [0]
y = [0]
n = 1000

for i in random.choices([0, 1, 2, 3], weights=[85, 7, 7, 1], k = n):
    x.append(array[i][0][0] * x[-1] + array[i][0][1] * y[-1] + array[i][0][2])
    y.append(array[i][1][0] * x[-2] + array[i][1][1] * y[-1] + array[i][1][2])
plt.plot(x, y, marker=".", markersize=0.3, linewidth=0)
plt.show()
```
