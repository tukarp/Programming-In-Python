### Paprotka Bransleya

### Importowanie bibliotek

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

### Funkcja generująca Paprotkę Bransleya

```python
# Funkcja generująca Paprotkę Bransleya
def generate_barnsleys_fern(n):
    # Zainicjalizowanie wartości w tablicy
    array = np.array([
        [[0.85, 0.04, 0], [-0.04, 0.85, 1.6]],
        [[-0.15, 0.28, 0], [0.26, 0.24, 0.44]],
        [[0.2, -0.26, 0], [0.23, 0.22, 1.6]],
        [[0, 0, 0], [0, 0.16, 0]]
    ])

    # Zainicjalizowanie wartości zmiennych
    x = [0]
    y = [0]

    # Obliczanie następnych wartości za pomocą wzoru
    for i in random.choices([0, 1, 2, 3], weights=[85, 7, 7, 1], k = n):
        x.append(array[i][0][0] * x[-1] + array[i][0][1] * y[-1] + array[i][0][2])
        y.append(array[i][1][0] * x[-2] + array[i][1][1] * y[-1] + array[i][1][2])

    # Rysowanie wykresu
    plt.plot(x, y, marker=".", markersize=1.5, linewidth=0)

    # Wyświetlenie wykresu
    plt.show()
```

### Generacja Paprotki Bransleya

Dane:

- n = 1 000

```python
# Generacja Paprotki Bransleya dla n = 1 000
generate_barnsleys_fern(1_000)
```

- n = 10 000

```python
# Generacja Paprotki Bransleya dla n = 10 000
generate_barnsleys_fern(10_000)
```

- n = 100 000

```python
# Generacja Paprotki Bransleya dla n = 100 000
generate_barnsleys_fern(100_000)
```
