### Program zwracający stolicę po podaniu nazwy kraju

```python
# Funkcja zwracająca stolicę po podaniu nazwy kraju
def get_capital_city(country_name):
    # Otwarcie połączenia i pobranie kodu strony
    website = get("https://pl.wikipedia.org/wiki/" + country_name)
    # Konwertowanie kodu strony na tekst
    website_text = website.text
    # Zamknięcie połączenia ze stroną
    website.close()
    # Zwrócenie stolicy podanego kraju po znalezieniu jej w kodzie strony
    return website_text.split("Stolica</a>")[1].split('href="/wiki/')[1].split('"')[0]

# Wyświetlenie stolicy podanego kraju
print(f"Stolica Polski to: {get_capital_city('Polska')}")  # Stolica Polski
print(f"Stolica Danii  to: {get_capital_city('Dania')}")   # Stolica Danii
print(f"Stolica Peru   to: {get_capital_city('Peru')}")    # Stolica Peru
```

### Program zwracający stolicę po podaniu nazwy kraju - czytanie z kodu strony po jednej linijce

```python
# Funkcja zwracająca stolicę po podaniu nazwy kraju
def get_capital_city_v2(country_name):
    # Otwarcie połączenia i pobranie kodu strony
    website = get("https://pl.wikipedia.org/wiki/" + country_name, stream=True)
    # Zaalokowanie trzech łańcuchów znakowych a, b, c, d do
    # Znalezienia linii kodu strony z nazwą stolicy podanego państwa
    a = ""
    b = ""
    c = ""
    result = ""
    # Pętla iterująca po liniach kodu strony
    for line in website.iter_lines():
        # Zamiana wartości zmiennych a, b, c, d oraz dekodowanie ich wartości z systemu UTF-8 do 
        # Przetransferowania linii kodu strony z nazwą stolicy podanego państwa
        a, b, c, result = (b, c, result, line.decode("UTF-8"))
        # Sprawdzenie czy linia d ma nazwe stolicy podanego państwa
        if "Stolica</a>" in a:
            # Zamknięcie połączenia ze stroną
            website.close()
            # Zwrócenie stolicy podanego kraju po znalezieniu jej w kodzie strony
            return result.split('">')[1].split("<")[0]

# Wyświetlenie stolicy podanego kraju
print(f"Stolica Włoch     to: {get_capital_city_v2('Włochy')}")    # Stolica Włoch
print(f"Stolica Nigerii   to: {get_capital_city_v2('Nigeria')}")   # Stolica Nigerii
print(f"Stolica Mozambiku to: {get_capital_city_v2('Mozambik')}")  # Stolica Mozambiku
```

### Sprawdzenie działania wyszukiwania stolic dla wielu państw jednocześnie

```python
# Lista z nazwami krajów
countries = [
    "Stany_Zjednoczone", "Niemcy", "Kanada", "Islandia", "Azerbejdżan",
    "Japonia", "Tanzania", "Timor_Wschodni", "Australia", "Jordania"
]

# Wylistowanie podanych państw do wyszukania stolicy
print(f"Lista państw: {list(countries)}")

# Zmapowanie i wylistowanie wyników wyszukiwania stolic podanych państw
print(f"Lista stolic: {list(map(get_capital_city, countries))}")
```

### Wykorzystanie wielowątkowości w języku Python

```python
# Funkcja zwracająca podaną stolicę z wykorzystaniem wielowątkowości
def thread_get_capital_city(country_name):
    print(f"Stolicą {country_name} jest {get_capital_city_v2(country_name)}")

# Lista z nazwami krajów
countries = [
    "Demokratyczna_Republika_Konga", "Kazachstan", "Irlandia", "Togo", "Wyspy_Owcze",
    "Korea_Północna", "Luksemburg", "Sierra_Leone", "Benin", "Andora"
]

# Utworzenie listy wątków dla podanych w liście państw
# Użycie funkcji thread_get_capital_city z utworzonymi wątkami
threads = [Thread(target=thread_get_capital_city, args=(country, )) for country in countries]

# Dla każdego wątku w liście wątków
for thread in threads:
    # Rozpocznij wątek
    thread.start()

# Dla każdego wątku w liście wątków
for thread in threads:
    # Dołącz wątek
    thread.join()
```

### Wyświetlanie podanego słowa z użyciem wielowątkowości

```python
# Funkcja wyświetlająca argument znak po znaku odczekując (0.5) sekundy między znakami
def old_print(* string):
    # Dla każdego znaku w łancuchu znaków
    for char in string:
        # Wyświetl znak
        print(char, end="")
        # Odczekaj pół sekundy
        sleep(0.5)
    # Odczekaj pół sekundy
    sleep(0.5)

# Wywołanie funkcji old_print() z podanym tekstem
old_print("Hello, World!")
```

### Wielowątkowość z użyciem zamka wątków

```python
# Funkcja wyświetlająca argument znak po znaku odczekując (0.5) sekundy między znakami
def old_print(* string):
    # Dla każdego znaku w łancuchu znaków
    for char in string:
        # Wyświetl znak
        print(char, end="")
        # Odczekaj pół sekundy
        sleep(0.5)
    # Odczekaj pół sekundy
    sleep(0.5)

# Funkcja wyświetlająca wyraz znak po znaku z użyciem wątków i zamka wątów
def thread_lock_print(string):
    # Dla każdego znaku w łancuchu znaków
    for char in range(string):
        # Z użyciem zamka wątków
        with lock:
            # Wyświetl znak z użyciem funkcji old_print()
            old_print(char)

# Utworzenie zamka wątków
lock = Lock()

# Utworzenie wątku pierwszego na rzecz funkcji thread_lock_print() z argumentem 10
thread_one = Thread(target=thread_lock_print, args=(10, ))
# Utworzenie wątku drugiego na rzecz funkcji thread_lock_print() z argumentem 10
thread_two = Thread(target=thread_lock_print, args=(10, ))

# Rozpoczęcie wątku pierwszego
thread_one.start()
# Rozpoczęcie wątku drugiego
thread_two.start()

# Dołączenie wątku pierwszego
thread_one.join()
# Dołączenie wątku drugiego
thread_two.join()
```
